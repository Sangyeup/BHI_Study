# PCA, t-SNE with BRCA dataset

## Load data

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.target
```

```python
data.target_names
```

```python
data.data.shape
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df['diagnosis'] = data.target
```

```python
df
```

```python
X = df.iloc[:, :30]
Y = df.iloc[:,30:]
```

```python
Y
```

## PCA

```python
# import PCA
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
pca_df = pd.DataFrame(X_pca, columns = ['PC1', 'PC2'])
pca_df
```

```python
pca_df['Diagnosis'] = Y
```

```python
pca_df
```

## Visualization

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'PC1', y = 'PC2', hue = 'Diagnosis', data=pca_df)
plt.title("PCA",fontsize=15)
plt.show()
```

## t-SNE

```python
# import t-SNE

from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, random_state = 0)

X_tsne = tsne.fit_transform(X)
```

```python
X_tsne.shape
```

```python
tsne_df = pd.DataFrame(X_tsne, columns = ['x', 'y'])
tsne_df['Diagnosis'] = Y
tsne_df
```

### Visualization

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'x', y = 'y', hue = 'Diagnosis', data=tsne_df)
plt.title("t-SNE",fontsize=15)
plt.show()
```
